Use the data at the following [API](https://pandas.pydata.org/docs/getting_started/index.html) to create a simple timesries plot of the temperature forecast (at 2 meters).

```python
import pandas as pd
import requests
import matplotlib.pyplot as plt

url = 'https://api.open-meteo.com/v1/forecast?latitude=52.52&longitude=13.41&current_weather=true&hourly=temperature_2m,relativehumidity_2m,windspeed_10m'
weather_r = requests.get(url)
data = weather_r.json()['hourly']
df = pd.DataFrame(data)
```

```python
df.set_index(df['time'], inplace = True)
df['temperature_2m'].plot(figsize = (15,9))
plt.ylim(bottom = 1)
plt.xticks(rotation = 80)
plt.show()
```
